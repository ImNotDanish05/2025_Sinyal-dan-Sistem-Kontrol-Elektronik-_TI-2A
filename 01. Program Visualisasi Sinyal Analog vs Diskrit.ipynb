```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
# Frekuensi gelombang (Hz)
frequency = 20 # @param
# jumlah titik diskrit
sampling_points = 100 # @param
# durasi sinyal dalam detik
duration = 1 # @param

# Data untuk sinyal analog (gelombang sinus)
t = np.linspace(0, duration, 500)  # waktu dalam detik
analog_signal = np.sin(2 * np.pi * frequency * t)

# Data untuk sinyal diskrit (sampling dari gelombang sinus)
t_discrete = np.linspace(0, duration, sampling_points)
discrete_signal = np.sin(2 * np.pi * frequency * t_discrete)

# Plot
plt.figure(figsize=(12, 6))

# Sinyal Analog
plt.subplot(2, 1, 1)
plt.plot(t, analog_signal, 'b')
plt.title(f"Sinyal Analog (f = {frequency} Hz, Durasi = {duration} s)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo")
plt.grid(True)

# Sinyal Diskrit
plt.subplot(2, 1, 2)
plt.stem(t_discrete, discrete_signal, basefmt=" ", linefmt="r-", markerfmt="ro")
plt.title(f"Sinyal Diskrit (f = {frequency} Hz, Sampling = {sampling_points} titik, Durasi = {duration} s)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo")
plt.grid(True)

plt.tight_layout()
plt.show()
```
